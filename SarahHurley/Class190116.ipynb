```python
import tellurium as te
# substrate S1 is converted to substrate S2
# first order rate law
# set k1 rate constant
# initialize S1
r = te.loada('S1 -> S2; k1*S1; k1 = 0.1; S1 = 10')
r.simulate(0, 50, 100) # from, to, steps
r.plot()
```

```python
import tellurium as te
te.setDefaultPlottingEngine('plotly')

model1 = """
model test
    compartment C1;
    C1 = 1.0;       # volume of the compartment
    species S1, S2;

    S1 = 10.0;
    S2 = 0.0;
    S1 in C1; S2 in C1;
    J1: S1 -> S2; k1*S1;

    k1 = 1.0e-03;
end
"""
# load models
r = te.loada(model1)

r.simulate(0, 50, 100) # from, to, steps
r.plot()
```

```python
import tellurium as te
te.setDefaultPlottingEngine('plotly')

# Production (P) = 2.45 
# But need this to be constant with time
# M0 constant, boundary condition
# const species S1
# M0 -> M1; K0*M0

# Decay(D)
# M1 -> M2; k1*M1

# Don't have to worry about signs in your rate constants because 

model1 = """
model test
    compartment C1;
    C1 = 1.0;       # volume of the compartment
    const species S0;
    var species S1, S2;

    S0 = 2.45
    S1 = 10.0;
    S2 = 0.0;
    S0 in C1; S1 in C1; S2 in C1;
    J0: S0 -> S1; k0*S0
    J1: S1 -> S2; k1*S1;

    k0 = 1.0;
    k1 = 1.209e-4;
end
"""
# load models
r = te.loada(model1)

r.simulate(0, 40000, 100, ["time","S1"]) # from, to, steps
r.plot()
```
