Simple model of radioactive decay.

This is an attempt understand how compartments work, using the well-known kinetics of radioactive decay. The compartment size is used to multiply the initial value assigned to the species to get a total absolute amount of the species that is present. However, it also looks like the timescale (including the half-life) of decay is proportional to the compartment size, which doesn't intuitively make sense.   

```python
import tellurium as te
```

```python
model RadDecayTest
    compartment C1;
    C1 = 1.0e0;
    
    species S1, S2;
    S1 = 1000.0 ; 
    S2 = 0.0 ;
    S3 = 1000.0 ;
    S4 = 0.0 ;

    S1 in C1; S2 in C1; S3 in C1; S4 in C1;

    
    half_life_U238 = 4.468e9 ; # half life of 238U in years
    k_decay_U238 = -1. * log(0.5) / half_life_U238
    
    J1: S1 -> S2; C1 * k_decay_U238 * S1
    J2: S3 -> S4; k_decay_U238 * S3
end
```

```python
RadDecayTest.reset()

start_time = 0.0
end_time = 4.468e9
time_steps = 100

RadDecayTest.simulate(start=start_time, end=end_time, steps=time_steps, selections=['time', "S1", "S2"]) 

RadDecayTest.plot()

print("half life (days)", end_time )
```
