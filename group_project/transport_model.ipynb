### transporty bits of model

```python
model c_transport

compartment env, cell;
    env = 1;
    cell = 1;
    
    species cxi in env, ci in cell;

vt = ktcat * T * (cxi / (kt + cxi)) * (1/ke); # transport rate
vm = kmcat * M * (ci / (km + ci)) * (1/ke); # assimilation rate

    cxi = 1000; #external inorg carbon, uM, pulled from figure 3 caption 
    T = 4.2e6; #transporter concentration
    M = 5e5; # metabolic enzyme concentration (used 0.06 as protein mass fraction) 
    ke = 10000; # half sat of e, molecs/cell 
    kt = 15; # half sat of T, uM. 
    km = 181; # half sat of M, uM. the table says 2441560 molecs/ cell so I converted to uM by dividing by the cell volume and avagadros number. I think this was the right move? still not sure why it was given in different units than kt though, so maybe I'm missing something.
    ktcat = 726; # min^-1, max import rate
    kmcat = 545; # min^-1, max metabolic rate 
    mc = 5; #average aa chain size, defined from table
    u = 0.06; #growth rate, u[h^-1] 
    
J1: cxi -> ci; vt - mc * vm - u * ci
    
end  
```

### plotting

```python
# simulate from 0 to 150 with 100 steps
c_transport.reset() #refreshes from the model
c_transport.simulate(start = 0, end = 150, steps = 100,
                selections = ['time', "cxi", "ci"]) # time is important for how it gets divied up
# plot the simulation
c_transport.plot()
```
