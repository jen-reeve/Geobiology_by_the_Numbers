```python
import tellurium as te
```

```python
model vm
    J0: ci + e + M -> aa + M; M * km_cat * (ci/(Km+ci)) * (e/(Ke+e));
    km_cat = 545;
    Km = 2441560;
    Ke = 10000;
end

model biosyn
    var ci;
    var e;
    var M;
    var aa;
    A: vm()
    A.ci is ci;
    A.e is e;
    A.M is M;
    A.aa is aa;
    aa = 0;
    M = 0.06*1.4e10 / 28630;
    ci = 2441560;
    e = 10000;
end
```

```python
biosyn.reset()
biosyn.simulate(start = 0, end = 1000, steps = 1000,selections =['time','e','ci','aa'])
biosyn.plot()
```
