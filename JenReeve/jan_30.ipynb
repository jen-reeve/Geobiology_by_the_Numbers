Lactose pathway model (copied last week's model)

```python
import tellurium as te
print("success")
```

```python
model = """
model test
    compartment cytoplasm;
    cytoplasm = 1.0; #this is a rather unitless volume, if you increase this you increase the volume and thus the reaction proceeds slower
    var species lactose, glucgalac;

    lactose = 10.0;
    glucgalac = 0.0;
    lactose in cytoplasm; glucgalac in cytoplasm;
    J1: lactose -> glucgalac; (kcat*lactose*Etot)/(km+lactose);

    kcat = 10.0;
    km = 5.5; #from Brenda
    Etot = 10.0;
end
"""
# load models
r = te.loada(model)
r.simulate(0,10,1000,["lactose","J1"])
r.plot(xtitle="Lactose (mM)",ytitle="V(S)")
```
