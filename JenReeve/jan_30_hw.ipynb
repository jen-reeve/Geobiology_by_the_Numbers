```python
model LacPath
    compartment C1;
    C1 = 1.0;
    species S1, S2;
    unit M = mole / liter
    unit inv_sec = 1. / seconds
    
    S1 = 10.e-3 M; # M; free parameter
    S2 = 0.0 M;

    S1 in C1; S2 in C1;
    
    J1: S1 -> S2; kcat*Etot*S1/(Km+S1);

    kcat = 6.42e2 inv_sec; #s^-1; from Juers et al (2012) 
    Etot = 1.e-6 M; # M; free parameter
    Km = 5.5e-3 M; # M; from BRENDA
end
```
